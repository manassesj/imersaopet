```python
#A coluna video_release_date está completamente vazia, então vai ser retirada
#Existe uma colunas com titulo unknown 

data_movie_copy = data_movie.copy()
data_movie_copy = data_movie_copy.drop(["video_release_date", "unknwon"], axis=1)
data_movie_copy.head()
```

```python
#Alguns IMDb_URL não possuem valor, mas como a url possue um padrão é possivel gerar resultados para ela
#Foi identificado que um dos filmes não possui valores, então vais ser excluido antes da opereção
data_movie_copy.loc[266] = ""
data_index_na_imbd_url = data_movie_copy.loc[data_movie_copy.isna().any(axis=1), "IMDb_URL"]
data_na_imbd_url = data_movie_copy.loc[data_index_na_imbd_url.index]
data_na_imbd_url.head()
```

```python
#Função que cria as urls com base no titulo para depois substituir no dataset
def build_url(title):
    base_url = "http://us.imdb.com/M/title-exact?"
    title_words = title.split(" ")
    
    for word in title_words[0:-1]:
        base_url += "{}%20".format(word)
    base_url += title_words[-1]
    return base_url

def fill_na_imbd_url(data):
    data_result = data.apply(lambda movie: build_url(title=movie["movie_title"]), axis=1)
    return data_result

data_imbd_result = fill_na_imbd_url(data_na_imbd_url)

#Substituindo os valores gerados nos dados
data_movie_copy.loc[data_movie_copy.isna().any(axis=1), "IMDb_URL"] = data_imbd_result

#Verificando as linhas que foram modificadas através dos indexes
indexes = data_imbd_result.index

data_movie_copy.loc[indexes]
```
