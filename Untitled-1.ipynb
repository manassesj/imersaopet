```python
#Alguns IMDb_URL nÃ£o possuem valor, mas como a url possue um padrÃ£o Ã© possivel gerar resultados para ela
#Foi identificado que um dos filmes nÃ£o possui valores, entÃ£o vais ser excluido antes da opereÃ§Ã£o
data_movie_copy.drop(266)
data_index_na_imbd_url = data_movie_copy.loc[data_movie_copy.isna().any(axis=1), "IMDb_URL"]
data_na_imbd_url = data_movie_copy.loc[data_index_na_imbd_url.index]
data_na_imbd_url.head()
```

```python
#FunÃ§Ã£o que cria as urls com base no titulo para depois substituir no dataset
def build_url(title):
    base_url = "http://us.imdb.com/M/title-exact?"
    title_words = title.split(" ")
    
    for word in title_words[0:-1]:
        base_url += "{}%20".format(word)
    base_url += title_words[-1]
    return base_url

def fill_na_imbd_url(data):
    data_result = data.apply(lambda movie: build_url(title=movie["movie_title"]), axis=1)
    return data_result

data_imbd_result = fill_na_imbd_url(data_na_imbd_url)
indexes = data_imbd_result.index

data_movie_copy.loc[data_movie_copy.isna().any(axis=1), "IMDb_URL"] = data_imbd_result
data_movie_copy.loc[indexes]
```
